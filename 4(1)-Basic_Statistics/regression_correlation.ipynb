```python
### 필요한 라이브러리 불러오기!
import pandas as pd
import statsmodels.api as sm
```

```python
### Advertising.csv를 불러와 데이터 로드하기!
data = pd.read_csv('Advertising.csv', index_col=0) #첫번째 data는 index로 처리
data.head()
```

```python
### Predictor Variable과 Response Variable을 설정해주세요!
X = data[['TV', 'radio', 'newspaper']]
y = data['sales']

# 상수항 추가 (Intercept를 포함하기 위해)
X = sm.add_constant(X)
X.rename(columns={'const': 'Intercept'}, inplace=True)
```

```python
### Multiple Linear Regression을 수행해봅시다!
model = sm.OLS(y, X).fit()
model.summary()
```

```python
### Correlation Matrix를 만들어 출력해주세요!
# 변수 간의 상관 계수 행렬 계산 및 출력
correlation_matrix = data.corr()
correlation_matrix
```
